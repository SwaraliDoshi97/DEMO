# **Heatmap**


```python
import numpy as np
```

```python
df=sns.load_dataset('anagrams')
df.head(2)
```

```python
df1=df.drop('attnr',axis=1)
df1.head(2)
```

```python
df1.corr()
```

```python
sns.heatmap(df1.corr())
```

```python
sns.heatmap(df1.corr(),cmap="PuOr")
```

```python
sns.heatmap(df1.corr(),annot=True)
```

```python
sns.heatmap(df1.corr(),annot=True,vmin=0,vmax=15,cmap='RdBu')
```

```python
sns.heatmap(df1.corr(),annot=True,vmin=0,vmax=15,cmap='RdBu',xticklabels=False,yticklabels=False)
plt.show()
```
